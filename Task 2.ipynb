```python
import pandas as pd

data = pd.read_csv('adult.data.csv', header=None, names=['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'salary'], skiprows=1)

data
```

```python
sex_counts = data['sex'].value_counts()
print(sex_counts)
```

```python
women_data = data[data['sex'] == 'Female']  # 取出性別為女性的資料
avg_women_age = women_data['age'].mean()    # 計算女性年齡的平均值
print("The average age of women is:", avg_women_age)
```

```python
# 計算每個國家的出現次數
country_counts = data['native-country'].value_counts()

# 獲取德國的出現次數
germany_count = country_counts['Germany']

# 計算德國公民在數據集中的佔比
percentage = germany_count / data.shape[0] * 100

print("德國公民在數據集中的佔比為 {:.2f}%".format(percentage))
```

```python
# 分組計算平均值和標準差
grouped = data.groupby('salary')['age'].agg(['mean', 'std'])
# 打印結果
print(grouped)
```

```python
# 分組計算平均值和標準差
grouped = data.groupby('salary')['age'].agg(['mean', 'std'])
# 打印結果
print(grouped)
```
