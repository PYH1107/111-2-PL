```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取数据
shooting = pd.read_csv('shooting.csv', parse_dates=['year'])

# 按州统计枪击案总数
state_counts = shooting['location'].value_counts()

# 取前 10 个州的数据
top_states = state_counts.head(10)

# 创建条形图
fig, ax = plt.subplots()
ax.bar(top_states.index, top_states.values)

# 添加标题和标签
ax.set_title('Gun Violence by State')
ax.set_xlabel('City')
ax.set_ylabel('Total Incidents')

# 自动旋转 X 轴标签
plt.xticks(rotation=90)

# 显示图形
plt.show()
```
